```python
user_profile = {
    "interests": ["AI", "Machine Learning"],
    "career_goal": "Become a Data Scientist",
    "skill_level": "Intermediate",
    "preferred_language": "English",
    "study_time_per_week": 5,
    "preferred_course_type": "Self-paced",
    "budget": "Free"
}
```

```python
filtered_df = df[
    (df["Category"].isin(user_profile["interests"])) |
    (df["Sub-Category"].isin(user_profile["interests"])) |
    (df["Skills"].str.contains("|".join(user_profile["interests"]), na=False)) &
    (df["Language"] == user_profile["preferred_language"])
]
```

```python
filtered_df = filtered_df[["Title", "URL", "Short Intro", "Skills", "Rating", "Number of viewers", "Duration"]]
```
